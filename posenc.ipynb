```python
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

```python
max_len = 128
d_model = 128
denom = 10000
embedding = nn.Embedding(max_len, d_model).requires_grad_(False)
pos = torch.arange(0.0, max_len).unsqueeze(-1)
div = torch.exp(-torch.arange(0, d_model, 2.0)) / d_model * torch.log(torch.tensor(denom))
embedding.weight[:, 0::2] = torch.sin(div * pos)
embedding.weight[:, 1::2] = torch.cos(div * pos)
```

```python
x = embedding.weight.data.numpy()
plt.figure(dpi=200)
plt.imshow(x)
plt.show()
```
