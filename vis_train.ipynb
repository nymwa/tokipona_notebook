```python
import numpy as np
import torch
import matplotlib.pyplot as plt
```

```python
with open('../../../mopumopu/220218/train.log') as f:
    log1 = [x.strip() for x in f]

with open('../../../mopumopu/220219/train.1.log') as f: # word dropout 0.2
    log2 = [x.strip() for x in f]
    
with open('../../../mopumopu/220219/train.log') as f: # word dropout 0.2
    log3 = [x.strip() for x in f]
    
def get_loss_list(log, name):
    lst = [x for x in log if '| {} |'.format(name) in x]
    lst = [x.split('|')[3] for x in lst]
    lst = [x.split()[1] for x in lst]
    lst = [float(x) for x in lst]
    return lst

t1 = get_loss_list(log1, 'train')
t2 = get_loss_list(log2, 'train')
t3 = get_loss_list(log3, 'train')
v1 = get_loss_list(log1, 'valid')
v2 = get_loss_list(log2, 'valid')
v3 = get_loss_list(log3, 'valid')

plt.plot(t2)
plt.plot(t3)
plt.show()

plt.plot(v2)
plt.plot(v3)
plt.show()

width = 30
for lst in [v2, v3]:
    plt.plot([np.mean(x) for x in zip(*[lst[i:] for i in range(width)])])
plt.show()
```
