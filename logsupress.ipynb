```python
import math
import numpy as np
import torch
import matplotlib.pyplot as plt
```

```python
lst = [n for n in range(2000)]
lst = [math.ceil(math.sqrt(n)) for n in lst]
for z in [10, 100, 1000]:
    xs = [n + 1 for n in range(z)]
    ys = [lst[x] for x in xs]
    plt.plot(xs, ys)
    plt.show()
```
